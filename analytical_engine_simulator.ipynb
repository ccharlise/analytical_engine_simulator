```python

# Analytical Engine Simulator
# Claudia Belk G01305636
# Created 01/26/2026, Last Edit 01/27/2026, Comments Added 02/21/2026

#### Functions ####

# factorial
def mill_factorial(number):
  if number <= 1:
    return 1
  return number * mill_factorial(number - 1)

# fibonacci
def mill_fibonacci(number1, number2, amount):
  fib = [number1, number2]
  for i in range(amount - 2):
    fib.append(fib[-1] + fib[-2])
  return fib

#### Main Code ####

# menu
while True:
  print("""-----------------------------
  Welcome to Analytical Engine Simulator!
  -----------------------------
  ------------Menu-------------
  1: Factorial
  2: Fibonacci
  3: Quit""")


  answer = int(input("Enter number: ")) # user inputs number from menu

  # factorial
  if answer == 1:
    print("You selected Factorial.")
    amount = int(input("Enter an amount of integers to store: "))
    # create store and output for the amount of integers
    store = [0] * amount
    output = [0] * amount

    for i in range(amount):
      store[i] = int(input("Import number: ")) # user inputs number into store
      print("-----------------------------")
      print("Store: ", store) # show store
      print("Calculating in mill...")
      output[i] = mill_factorial(store[i]) # compute using factorial function
      print("Output: ", output) # show result
      print("-----------------------------")

  # fibonacci
  elif answer == 2:
    print("You selected Fibonacci.")
    amount = int(input("Enter an amount of fibonacci sequence to store: "))
    # create store in the size of amount of fibonacci sequence
    store = [0] * 2

    for i in range(2):
      store[i] = int(input("Import number: ")) # user inputs number into store
      print("Store: ", store) # show store
      print("-----------------------------")

    print("Calculating in mill...")
    output = mill_fibonacci(store[0], store[1], amount) # compute using fibonacci function
    print("-----------------------------")
    print("Output: ", output) # show result

  # quit
  elif answer == 3:
    print("""Quiting...
    Thank you for using Analytical Engine Simulator!""")
    break # ends run

  # invalid input
  else:
    print("Please enter numbers 1-3.") # remind user of correct input options


```
